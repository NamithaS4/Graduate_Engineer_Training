```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
arr = np.array([10, 20, 30])
df = pd.DataFrame({"Numbers": arr})
plt.bar(df.index, df["Numbers"])
plt.xlabel("Index")
plt.ylabel("Numbers")
plt.title("Bar Plot from DataFrame")
plt.xticks(df.index, ['A', 'B', 'C'])
plt.show()
```

##if we need only one column then we should use df.column_name.quantile

```python
Q1 = df.MRP.quantile(0.25)
Q3 = df.MRP.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

##if the data contains values of other than numeric then we neede to use "numeric_only = true"

```python
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```
